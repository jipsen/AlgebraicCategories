```python
import os, re

def replf(matchobj):
    return '['+matchobj.group(1)+']('+matchobj.group(1).lower().replace(' ','_')+'.md)'

def replfs(matchobj):
    return '['+matchobj.group(1)+']('+matchobj.group(1).lower().replace(' ','_')+('.md)' if matchobj.group(1)[-1] == 's' else 's.md)')

def find_and_replace_in_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    
    content = re.sub(r'=====(.*?)=====', r'# \1', content)
    content = re.sub(r'====(.*?)====', r'## \1', content)
    content = re.sub(r'==(.*?)==', r'### \1', content)
    content = re.sub(r'\\emph\{(.*?)\}', r'***\1***', content)
    content = re.sub(r'\\textbf\{(.*?)\}', r'**\1**', content)
    content = content.replace(r'^[[Classtype', '\n\n|Property|Value|\n|---|---|\n|[[Classtype')
    content = content.replace(r'^[[', r'|[[')
    #content = content.replace(r'[[', r'<<')
    #content = content.replace(r']]', r'>>')
    content = re.sub(r'\|\[\[(.*?)\]\]', replf, content)
    content = re.sub(r'\[\[(.*?)\]\]', replfs, content)

    with open(file_path, 'w') as file:
        file.write(content)

def process_folder(folder_path):
    i = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('abelian_ordered_groups.md'):
                file_path = os.path.join(root, file)
                find_and_replace_in_file(file_path)
                i += 1
    print(f'Processed {i} files')

# Example usage
folder_path = '.'

process_folder(folder_path)
```
