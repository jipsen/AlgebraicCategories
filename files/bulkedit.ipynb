```python
import os, re

def nextmath(st,i): #find next j,k>=i such that st[j:k] is inline math
    j = st.find("$",i)
    if j==-1: return (-1,0)
    return (j+1,st.find("$",j+1))

def checkmath(st):
    i = 0
    while i<len(st):
        j,k = nextmath(st,i)
        if j==-1: return True
        if k==0:
            print("Unmatched $ in",st)
            return True
        if st[j+1]==" ":
            print("Space after $")
        if st[k-1]==" ":
            print("Space before $")
        i = k+1
    return False

def check_file(file_path,i):
    with open(file_path, 'r') as file:
        content = file.read()
    flag = False
    if re.search(r' \$ ', content):
        print(i+1,' $ ',file_path)
        flag = True
    matchobj=re.search(r'\\emph\{.*?\}', content)
    if matchobj and re.search(r'\$', matchobj.group(0)):
        print(i+1,'{$}',file_path)
        flag = True
    if re.search(r'\$\$', content):
        print(i+1,'$$',file_path)
        flag = True
    if not checkmath(content):
        print(i+1,'spc',file_path)
    return i+1 if flag else i   

def fixmath(st):
    i = 0
    while i<len(st):
        j,k = nextmath(st,i)
        if j==-1: return st
        if k==0:
            print("Unmatched $ in",st)
            return st
        if st[j+1]==" ":
            st = st[:j+1]+st[j+2:]
        if st[k-1]==" ":
            st = st[:k-1]+st[k:]
        i = k+1
    return st

def replf(matchobj):
    return '|['+matchobj.group(1)+']('+matchobj.group(1).lower().replace(' ','_')+'.md)'

#def replf2(matchobj):
#    return '['+matchobj.group(1)+']('+matchobj.group(1).lower().replace(' ','_')+('.md)' if matchobj.group(1)[-1] == 's' else 's.md)')

def replfs(matchobj):
    return '['+matchobj.group(1)+']('+matchobj.group(1).lower().replace(' ','_')+('.md)' if matchobj.group(1)[-1] == 's' else 's.md)')

def find_and_replace_in_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    
    content = re.sub(r'=====(.*?)=====', r'# \1', content)
    content = re.sub(r'====(.*?)====', r'## \1', content)
    content = re.sub(r'==(.*?)==', r'### \1', content)
    content = re.sub(r'\\emph\{(.*?)\}', r'***\1***', content)
    content = re.sub(r'\\textbf\{(.*?)\}', r'**\1**', content)
    content = re.sub(r'\\mbox\{(.*?)\}', r'\text{\1}', content)
    content = content.replace(r'^[[Classtype', '\n\n|Property|Value|\n|---|---|\n|[[Classtype')
    content = content.replace(r'^[[', r'|[[')
    content = re.sub(r'\|\[\[(.*?)\]\]', replf, content)
    #content = re.sub(r'\[\[(.*?)\|(.*?)\]\]', replf2, content)
    content = re.sub(r'\[\[(.*?)\]\]', replfs, content)
    content = re.sub(r'\[\((.*?)\)\]', r'[\1]', content)
    content = re.sub(r'\[\((.*?)>', '', content)
    content = re.sub(r'\)\]\n', '\n', content)
    content = fixmath(content)
    with open(file_path, 'w') as file:
        file.write(content)

def process_folder(folder_path):
    i = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('commutative_monoids.md'):
                file_path = os.path.join(root, file)
                find_and_replace_in_file(file_path)
                i += 1
    print(f'Processed {i} files')

def check_folder(folder_path):
    i = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                i=check_file(file_path,i)
    print(f'Processed {i} files')

# Example usage
folder_path = '.'

process_folder(folder_path)
```
